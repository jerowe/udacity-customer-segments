### Implementation: Outlier Detection
Detecting outliers in the data is extremely important in the data preprocessing step of any analysis. The presence of outliers can often skew results which take into consideration these data points. There are many "rules of thumb" for what constitutes an outlier in a dataset. Here, we will use [Tukey's Method for identfying outliers](http://datapigtechnologies.com/blog/index.php/highlighting-outliers-in-your-data-with-the-tukey-method/): An *outlier step* is calculated as 1.5 times the interquartile range (IQR). A data point with a feature that is beyond an outlier step outside of the IQR for that feature is considered abnormal.

In the code block below, you will need to implement the following:
 - Assign the value of the 25th percentile for the given feature to `Q1`. Use `np.percentile` for this.
 - Assign the value of the 75th percentile for the given feature to `Q3`. Again, use `np.percentile`.
 - Assign the calculation of an outlier step for the given feature to `step`.
 - Optionally remove data points from the dataset by adding indices to the `outliers` list.

**NOTE:** If you choose to remove any outliers, ensure that the sample data does not contain any of these points!  
Once you have performed this implementation, the dataset will be stored in the variable `good_data`.

```python
# For each feature find the data points with extreme high or low values

outliers  = []
feature_outlier_bucket = {}
for feature in log_data.keys():
    
    # TODO: Calculate Q1 (25th percentile of the data) for the given feature
    Q1 = np.percentile(log_data[feature], 25.)
    
    # TODO: Calculate Q3 (75th percentile of the data) for the given feature
    Q3 = np.percentile(log_data[feature], 75.)
    
    # TODO: Use the interquartile range to calculate an outlier step (1.5 times the interquartile range)
    step = (Q3-Q1)*1.5
    
    # Display the outliers
    print("Data points considered outliers for the feature '{}':".format(feature))
    feature_outlier = log_data[~((log_data[feature] >= Q1 - step) & (log_data[feature] <= Q3 + step))]
    
    display(feature_outlier)
    
    feature_outlier_index_list = feature_outlier.index.tolist()
    feature_outlier_bucket[feature] = (feature_outlier_index_list)
    outliers += feature_outlier_index_list
    

# Remove the outliers, if any were specified
print('Number of outliers: {}'.format(len(outliers)))
print('Total number of data points {}'.format(log_data.shape[0]))
good_data = log_data.drop(log_data.index[outliers]).reset_index(drop = True)
```

```python
# OPTIONAL: Select the indices for data points you wish to remove
# https://stackoverflow.com/questions/3852780/python-intersection-of-multiple-lists
overlap_results = {}
for k1 in feature_outlier_bucket:
    for k2 in feature_outlier_bucket:
        if not k1 == k2:
            t_title = [k1, k2]
            t_title.sort()
            title = ' '.join(t_title)
            if not title in overlap_results:
                overlap_results[title] = list(set(feature_outlier_bucket[k1]) & set(feature_outlier_bucket[k2]))

print("Feature Sets with Overlap in Outliers\n")
overlap_indexes = []
for outlier_overlap in overlap_results:
    if len(overlap_results[outlier_overlap]):
        print('{}: {}'.format(outlier_overlap, overlap_results[outlier_overlap]))
        for row_index in overlap_results[outlier_overlap]:
            overlap_indexes.append(row_index)

overlap_indexes = list(set(overlap_indexes))          
print("Outlier Rows\n")
print(log_data.loc[overlap_indexes])
```

When the number of clusters is not known *a priori*, there is no guarantee that a given number of clusters best segments the data, since it is unclear what structure exists in the data â€” if any. However, we can quantify the "goodness" of a clustering by calculating each data point's *silhouette coefficient*. The [silhouette coefficient](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html) for a data point measures how similar it is to its assigned cluster from -1 (dissimilar) to 1 (similar). Calculating the *mean* silhouette coefficient provides for a simple scoring method of a given clustering.

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score


#https://scikit-learn.org/stable/modules/generated/sklearn.mixture.GaussianMixture.html
component_with_max_score = 1
max_score = 0
centers = 0
sample_preds = 0
preds = 0
clusterer = 0

# TODO: Apply your clustering algorithm of choice to the reduced data 
for n_components in [2,3,4,5,6]:
    t_clusterer = GaussianMixture(n_components=n_components)
    t_clusterer.fit(reduced_data)

    # TODO: Predict the cluster for each data point
    t_preds = t_clusterer.predict(reduced_data)

    # TODO: Find the cluster centers
    t_centers = t_clusterer.means_ 

    # TODO: Predict the cluster for each transformed sample data point
    t_sample_preds = t_clusterer.predict(pca_samples)


    # TODO: Calculate the mean silhouette coefficient for the number of clusters chosen
    score = silhouette_score(reduced_data, t_preds)
    if score >= max_score:
        max_score = score
        component_with_max_score = n_components
        centers = t_centers
        sample_preds = t_sample_preds
        preds = t_preds
        clusterer = t_clusterer
    
    print('N Components: {} Score: {}'.format(n_components, score))
    
print('Max Score: {}'.format(max_score))
print('N Component with Max Score {}'.format(component_with_max_score))
print('Centers: {}'.format(centers.shape))
```
